```python
import soditools.soditools as u
import pandas as pd
```

```python
_file = f'UPSELL_CO.csv'
_bucket = f'bucket_recommender_co'
_path = f'gs://{_bucket}/'

# df = pd.read_csv(_path+_file)
df = u.gcs_data_to_dataframe(_path+_file)
```

```python
df.head()
```

```python
df = df[['productId','upsell_a_sin_segmento']].copy()
df.columns = ['upsell','up_list']
df['upsell'] = df['upsell'].apply(lambda x: f'Upsell_{x}')
```

```python
df_up = df['up_list'].str.split("|", expand=True).copy()
df_up.columns = [f"Sku{i+1}" for i in range(df_up.shape[1])]
```

```python
df_up.head()
```

```python
df_final = pd.concat([df['upsell'], df_up], axis=1)
```

```python
df_final.sample(5).T
```

```python
df_final.info()
```

```python
del df, df_up
```

```python
chunks = u.np.array_split(df_final, 3)
```

```python
for i,p in enumerate(chunks):
    _file_chunk = f"{_file.split('.')[0]}_{i+1}.csv"
    p.to_csv(_file_chunk, index=False, header=True)

    u.transfer_local_file_to_gs(
        bucket_name=_bucket,
        destination_file=_file_chunk,
        origin_file=_file_chunk,
        project='sod-corp-plp-beta'
    )
```
